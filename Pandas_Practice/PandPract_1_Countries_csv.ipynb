```python
#Q:Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".
Highest_cases_df=combined_df.sort_values(by="cases_per_million",ascending=False)[:10]
Highest_tests_df=combined_df.sort_values(by="tests_per_million",ascending=False)[:10]
countries_in_both=(set(Highest_cases_df["location"]) & set(Highest_tests_df["location"]))
print(countries_in_both)
len(countries_in_both)
```
